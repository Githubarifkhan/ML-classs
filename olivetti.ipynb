```python
from sklearn.decomposition import PCA 
pca = PCA(2) 
pca
```

```python
fig = plt.figure(figsize=(8,8))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05) 
# plot the faces each image is 64 by 64 pixels
for i in range(64): 
    ax = fig.add_subplot(8, 8, i+1, xticks=[], yticks=[])
    ax.imshow(olive.images[i], cmap=plt.cm.bone, interpolation='nearest')
plt.show()
```

```python
X,y=olive.data,olive.target
pca_olive = PCA(64) 
X_proj = pca_olive.fit_transform(X)
X_proj.shape
```

```python
df =np.cumsum(pca_olive.explained_variance_ratio_)
df
```

```python
for i in range(10):
    ax = fig.add_subplot(5, 5, i+1, xticks=[], yticks=[])
    ax.imshow(np.reshape(pca_olive.components_[i,:], (64,64)),cmap=plt.cm.bone, interpolation='nearest')
```

```python
X_inv_proj = pca_olive.inverse_transform(X_proj)
X_proj_img = np.reshape(X_inv_proj,(400,64,64))
fig = plt.figure(figsize=(8,8))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.02, wspace=0.05)
for i in range(64):
    ax = fig.add_subplot(8, 8, i+1, xticks=[], yticks=[])
    ax.imshow(X_proj_img[i], cmap=plt.cm.bone, interpolation='nearest')
```

```python
from sklearn.decomposition import PCA as RandomizedPCA
```

```python
rpca_olive = RandomizedPCA(64).fit(X)
print("Randomized PCA with 64 components: ", np.sum(rpca_olive.explained_variance_ratio_))
print("PCA with 64 components: ", np.sum(pca_olive.explained_variance_ratio_))
```

```python
fig = plt.figure(figsize=(8,8))
fig.subplots_adjust(left=0, right=1, bottom=0, top=1, hspace=0.05, wspace=0.05)
for i in range(64): 
    ax = fig.add_subplot(8, 8, i+1, xticks=[], yticks=[]) 
    ax.imshow(np.reshape(rpca_olive.components_[i,:], (64,64)), cmap=plt.cm.bone, interpolation='nearest')
```
