# naive bayes on titanic dataet

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import time
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB, BernoulliNB,MultinomialNB
```

```python
data =pd.read_csv("Titanic_x_y_train.csv")
```

```python
data.head()
```

```python
#convert categorical variable into numeric
data ["Sex_cleaned"]=np.where(data["Sex"]=="male",0,1)

data["Embarked_cleaned"] =np.where(data["Embarked"] =="S",0,
                                  np.where(data["Embarked"] =="C",1,
                                          np.where(data["Embarked"]=="Q",2,3)))
```

```python
#cleaning dataset of NAN
data =data[["Survived", "Pclass", "Sex_cleaned", "Age", "SibSp", "Parch","Fare","Embarked_cleaned"]].dropna(axis =0,how ='any')
```

```python
#split dataset into training and testing
X_train,X_test =train_test_split(data,test_size =0.5,random_state=int(time.time()))
```

```python
gnb=GaussianNB()
used_features=["Pclass", "Sex_cleaned", "Age", "SibSp", "Parch","Fare","Embarked_cleaned"]
```

```python
#train classifier
gnb.fit(X_train[used_features].values,
       X_train["Survived"])
y_pred =gnb.predict(X_test[used_features])
```

```python
#print results
print("Number of mislabeled points out of a label{} points: {},perfomance{:05.2f}%".format(X_test.shape[0],(X_test["Survived"] !=y_pred).sum(),100*(1-(X_test["Survived"]!=y_pred).sum()/X_test.shape[0])
))
```

```python
#calculate the probability
mean_survival =np.mean(X_train["Survived"])
mean_not_survival =1-mean_survival
print("Survival prop ={:03.2f}%,Not survival prob ={:03.2f}%" .format(100*mean_survival ,100* mean_not_survival))
```
